```python
from transformers import pipeline


# Load models
generator = pipeline("text-generation", model="gpt2")
sentiment = pipeline("sentiment-analysis")

# Input
headline = "India launches new AI policy"

# Generate article
generated = generator(
    headline,
    max_length=120,
    temperature=0.8,
    do_sample=True,
    num_return_sequences=1
)

generated_news = generated[0]["generated_text"]

print("Generated News:\n")
print(generated_news)

# Sentiment Analysis
prediction = sentiment(generated_news)

print("\nSentiment Prediction:")
print("Label :", prediction[0]["label"])
print("Confidence :", prediction[0]["score"])
```
